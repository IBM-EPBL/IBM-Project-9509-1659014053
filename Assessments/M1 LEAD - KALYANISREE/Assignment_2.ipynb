**FIND THE OUTLIERS AND REPLACE THE OUTLIERS**

```python
sns.boxplot(x=df.CreditScore)
```

```python
q1 = df.CreditScore.quantile(0.25)
q3 = df.CreditScore.quantile(0.75)
IQR = q3 - q1
IQR
UpperFence = q3 + (1.5 * IQR)
UpperFence
LowerFence=q1-(1.5*IQR)
LowerFence
clist=df.CreditScore.to_list()
outlier=[]
for i in clist:
  if i<LowerFence:
    outlier.append(i)
outlier
```

```python
df.median(numeric_only=True)
```

```python
df.CreditScore.quantile(0.50)
```

```python
df1=df.copy()
df1['CreditScore'] = np.where(df['CreditScore'] < LowerFence, 652, df['CreditScore'])
df1.describe()
```

```python
# outliers removed
# sns.boxplot(x=df1.CreditScore)
df['CreditScore'] = np.where(df['CreditScore'] < LowerFence, 652, df['CreditScore'])
df.describe()
```
